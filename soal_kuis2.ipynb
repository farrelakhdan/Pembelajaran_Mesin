```python
import numpy as np
import pandas as pd

df = pd.read_excel('data/dataset_quiz.xlsx')

df.head(100)
```

```python
#PENGECEKAN NULL
#TERDAPAT BEBERAPA KOLOM YANG BERISI NULL
df.isnull().sum()
```

```python
#PEMBERIAN DATA YANG NULL DENGAN NILAI MODUS, MEAN dengan simple imputer
from sklearn.impute import SimpleImputer

pengisian = SimpleImputer(missing_values=np.nan, strategy="constant", copy=False)
fitur_kolom = ['id_ketua', 'id_anggota1', 'id_anggota2', 'id_anggota3', 'id_anggota4', 'id_anggota5']

df[fitur_kolom] = pengisian.fit_transform(df[fitur_kolom])
df.head()
```

```python
#PENGECEKAN NULL
#SUDAH RIDAK TERDAPAT KOLOM YANG BERISI NULL
df.isnull().sum()
```

SOAL NO 1

```python
from sklearn.model_selection import train_test_split

df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=3, stratify=df['keterlibatan_vltr'])

df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=30, stratify=df_unseen['keterlibatan_vltr'])
```

SOAL NO 2

```python
# 0 = YA
# 1 = TIDAK
# Terdapat 4 data validasi bernilai YA dan 4 data validasi berniali TIDAK
print(f'Label validasi : \n{df_val.keterlibatan_vltr.value_counts()}')
```

SOAL NO 3

```python
# import label encoder
# labeling manual
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

df['keterlibatan_vltr'] = label_encoder.fit_transform(df['keterlibatan_vltr'])
df['skema'] = label_encoder.fit_transform(df['skema'])
df['luaran_hki'] = label_encoder.fit_transform(df['luaran_hki'])
df['jenis_luaran_hki'] = label_encoder.fit_transform(df['jenis_luaran_hki'])
df['potensi_hki'] = label_encoder.fit_transform(df['potensi_hki'])
df['luaran_lain'] = label_encoder.fit_transform(df['luaran_lain'])
df['jenis_luaran_lain'] = label_encoder.fit_transform(df['jenis_luaran_lain'])
df['status_luaran_lain'] = label_encoder.fit_transform(df['status_luaran_lain'])
```

SOAL NO 4

```python
#SELEKSI FITUR 
X = df.iloc[:,0:16]
y = df['tahun']
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

KLASIFIKASI DENGAN DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

dt = DecisionTreeClassifier()

dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

acc_dt = accuracy_score(y_test, y_pred_dt)
print(f"Hasil score Decision Tree : {acc_dt}")
```

KLASIFIKASI DENGAN SVM POLYNOMIAL

```python
from sklearn.svm import SVC

svm = SVC(kernel='poly', degree=10)

svm.fit(X_train, y_train)

y_pred_svm = svm.predict(X_test)

score_svm = accuracy_score(y_test, y_pred_svm)

print(f"Hasil score SVM : {score_svm}")
```

SOAL NO 5

```python
from sklearn.model_selection import RepeatedKFold

cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=30)
```

GRID SEARCH

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 2)
k_range = list(range(1,8))
param_grid = dict(n_neighbors=k_range)

grid = GridSearchCV(knn, param_grid, cv=10, scoring='accuracy', return_train_score=False,verbose=1)

grid_search=grid.fit(X_train, y_train)

accuracy = grid_search.best_score_ *100
print("Akurasi dengan GridSearch : {:.2f}%".format(accuracy) )
```

RANDOMIZED SEARCH

```python
hyperparameter_space = {'max_depth':[2,3,4,6,8,10,12,15,20], 
                        'min_samples_split':[1,2,3,4,5,6,8,10]}

from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(dt, param_distributions=hyperparameter_space,
                        n_iter=10, scoring="accuracy", random_state=0,
                        n_jobs=-1, cv=10, return_train_score=True)

rs.fit(X_train, y_train)

rs.best_estimator_.fit(X_train, y_train)
y_pred = rs.best_estimator_.predict(X_test) 
y_true = y_test 

print("Akurasi dengan RandomizedSearch:", np.round(accuracy_score(y_true, y_pred), 2))
```
